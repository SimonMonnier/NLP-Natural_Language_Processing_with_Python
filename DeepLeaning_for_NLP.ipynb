Text Generation with keras

```python
def read_file(filepath):
    with open(filepath, encoding="utf8") as f:
        str_text = f.read()
    return str_text
```

```python
print(read_file('coran.txt'))
```

```python
import spacy
```

```python
nlp = spacy.load("fr_core_news_sm", disable=['parser','tagger','ner'])
```

```python
nlp.max_length = 50000000
```

```python
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
d = read_file('coran.txt')
```

```python
tokens = separate_punc(d)
```

```python
len(tokens)
```

```python
#50 words --> network predict #26
```

```python
train_len = 25 + 1

text_sequences = []

for i in range(train_len, len(tokens)):
    seq = tokens[i-train_len:i]
    
    text_sequences.append(seq)
```

```python
type(text_sequences)
```

```python
 from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
vocabulary_size = len(tokenizer.word_counts)
```

```python
import numpy as np

sequences = np.array(sequences)

sequences
```

```python
from keras.utils import to_categorical
```

```python
X = sequences[:,:-1]
```

```python
y = sequences[:,-1]
```

```python
y = to_categorical(y,num_classes=vocabulary_size+1)
```

```python
seq_len = X.shape[1]
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
def create_model(vocabulary_size,seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size,seq_len,input_length=seq_len))
    model.add(LSTM(50,return_sequences=True))
    model.add(LSTM(50))
    model.add(Dense(50,activation='relu'))
    model.add(Dense(vocabulary_size,activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    
    model.summary()
    
    return model
```

```python
model = create_model(vocabulary_size+1,seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(X[0:30000],y[0:30000],batch_size=128,epochs=300,verbose=2)
```

```python
model.save('coran_model.h5')
```

```python
from keras_preprocessing.sequence import pad_sequences
```

```python
def generate_text(model,tokenizer,seq_len,seed_text,num_gen_words):
    
    output_text = []
    
    input_text = seed_text
    
    for i in range(num_gen_words):
        
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        
        pad_encoded = pad_sequences([encoded_text],maxlen=seq_len,truncating='pre')
    
        pred_word_ind = model.predict(pad_encoded,verbose=0)[0]
 
        classes_x_ind =np.argmax(pred_word_ind)
        
        pred_word = tokenizer.index_word[classes_x_ind]
        
        input_text += ' ' + pred_word
        
        output_text.append(pred_word)
    
    return ' '.join(output_text)
```
